### grouping on month year basis

```python
month_year_wise = df.groupby('month_year')['Total Amount'].sum()
month_year_wise_sorted = month_year_wise.sort_values(ascending=False).reset_index()
```

```python
month_year_wise
```

```python
month_year_wise_sorted
```

```python
plt.plot(month_year_wise)
plt.title('month_year_wise__sales')
plt.xlabel('month_year')
plt.ylabel('Total Amount')
plt.grid(True)

plt.plot(month_year_wise, color='green',linestyle = 'solid', marker = 'o',markerfacecolor = 'red')

plt.xticks(rotation=45)
plt.tight_layout()
```

```python
month_year_wise.plot.bar()
```
